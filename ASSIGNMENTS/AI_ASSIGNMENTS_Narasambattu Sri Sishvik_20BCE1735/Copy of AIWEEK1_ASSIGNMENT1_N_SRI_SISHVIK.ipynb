```python
#Assignment 1 starts
#1st task- Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)
import numpy as npy
import pandas as pd
```

```python
#Set random seed for reproducing
npy.random.seed(100)
```

```python
#Assign numpy random values(4 features and 4 observations)
d=npy.random.randn(4,4)
```

```python
#Task-2: Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4'
df=pd.DataFrame(d, columns=['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4'])
print(df)
```

```python
#Task-3: Find descriptive statistics of df
print(df.describe())
```

```python
#Task-4: Check for the null values in 'df' and find the data type of the columns
null_check = df.isnull().sum()
d_types = df.dtypes
print("Null values:") 
print(null_check)
print("\nData Types:")
print(d_types)
```

```python
#Task-5: Display the 'Random value 2' & 'Random value 3' columns with location method and index location method.
loc_meth_cols=df.loc[:,['Random value 2', 'Random value 3']]
print("Location Method:")
print(loc_meth_cols)
iloc_meth_cols=df.iloc[:,[1,2]]
print("\nIndex Location Method:")
print(iloc_meth_cols)
```
